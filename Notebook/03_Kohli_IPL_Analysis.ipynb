# üèè Kohli IPL Performance Analysis

This notebook provides an in-depth analysis of Virat Kohli's performance in the **Indian Premier League (IPL)** over the years.

We will explore:
- Total runs per season
- Batting average and strike rate
- Number of 50s and 100s
- Consistency and standout seasons


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style for all plots
sns.set(style='whitegrid')
```

```python
df_ipl = pd.read_csv('../Data/Virat_Kohli_YearWise_with_IPL.csv')
df_ipl.head()
```

```python
df_ipl.columns
```

```python
df_ipl.columns = df_ipl.columns.str.strip().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_ipl, x='Year', y='Runs', palette='crest')
plt.title("Virat Kohli - Total IPL Runs Per Year")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## üìà Kohli's IPL Batting Average Over the Years
Let's explore how consistent Kohli has been in terms of his batting average throughout his IPL journey.


```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_ipl, x='Year', y='Average', marker='o', color='green')
plt.title("Virat Kohli - IPL Batting Average Per Year")
plt.xlabel("Year")
plt.ylabel("Batting Average")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## üß® Fifties and Centuries Comparison (Year-wise)
This chart compares the number of 50s and 100s Kohli scored in each IPL season.


```python
plt.figure(figsize=(10, 6))
plt.plot(df_ipl['Year'], df_ipl['Fifties'], marker='o', label='Fifties', color='blue')
plt.plot(df_ipl['Year'], df_ipl['Centuries'], marker='o', label='Centuries', color='red')
plt.title('Virat Kohli - 50s vs 100s Per IPL Season')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## üöÄ Kohli's Total IPL Runs Per Season
This chart highlights Kohli‚Äôs overall contribution to RCB each year by showing total runs scored per IPL season.


```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_ipl, x='Year', y='Runs', palette='viridis')
plt.title("Virat Kohli - Total IPL Runs Per Season")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## üíØ Kohli's Best Season in IPL - 2016
Kohli had a legendary IPL 2016 season, scoring **973 runs** ‚Äî the most ever in a single season! Let's highlight that.


```python
best_season = df_ipl[df_ipl['Runs'] == df_ipl['Runs'].max()]
print(best_season)
```

## üéØ Strike Rate Trend - Kohli's Attacking Style Over the Years
This chart reveals how Kohli‚Äôs batting approach evolved by tracking his strike rate each season.

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_ipl, x='Year', y='Matches', marker='o', label='Matches', color='blue')
sns.lineplot(data=df_ipl, x='Year', y='Innings', marker='o', label='Innings', color='orange')
plt.title("Virat Kohli - Matches vs Innings per IPL Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
